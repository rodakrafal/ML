```python
import numpy as np
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.base import clone

from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.linear_model import Lasso, Ridge, ElasticNet

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

X = df_combined  
# X = df_combined_only_sum
# X = df_combined_max
y = df_y_combined.values.ravel()

n_components = 2
# Set up the cross-validation scheme suitable for regression tasks.
# RepeatedKFold repeats K-Fold n times with different randomization in each repetition.
# cross-validation -> https://aszokalski.github.io/AI/Cross-Validation.html
# https://scikit-learn.org/stable/modules/cross_validation.html#cross-validation
rskf = RepeatedKFold(n_splits=2, n_repeats=5, random_state=42)

# Define base models
# Regression -> https://scikit-learn.org/stable/modules/svm.html#regression
base_models = [
    KNeighborsRegressor(n_neighbors=3),
    KNeighborsRegressor(n_neighbors=7),
    SVR(),
    DecisionTreeRegressor(random_state=41),
    RandomForestRegressor(random_state=41),
    GradientBoostingRegressor(random_state=41),
    AdaBoostRegressor(random_state=41),
    Lasso(),
    Ridge(),
    ElasticNet(),
]

# Define the model names
model_names = [
    'KNeighborsRegressor (n_neighbors=3)',
    'KNeighborsRegressor (n_neighbors=7)',
    'SVR',
    'DecisionTreeRegressor',
    'RandomForestRegressor',
    'GradientBoostingRegressor',
    'AdaBoostRegressor',
    'Lasso',
    'Ridge',
    'ElasticNet'
]

# Initialize arrays to store the scores for base and PCA models
base_metrics_scores = np.zeros((len(base_models), 3))  # 3 for R^2, MSE, and MAE
pca_metrics_scores = np.zeros((len(base_models), 3))  # 3 for R^2, MSE, and MAE

# Evaluate base models
for est_idx, est in enumerate(base_models):
    all_fold_r2 = []
    all_fold_mse = []
    all_fold_mae = []
    for train_idx, test_idx in rskf.split(X, y):
        clf = clone(est)
        clf.fit(X.iloc[train_idx], y[train_idx])
        y_pred = clf.predict(X.iloc[test_idx])

        # Metrics -> https://scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics
        all_fold_r2.append(r2_score(y[test_idx], y_pred))
        all_fold_mse.append(mean_squared_error(y[test_idx], y_pred))
        all_fold_mae.append(mean_absolute_error(y[test_idx], y_pred))

    base_metrics_scores[est_idx, 0] = np.mean(all_fold_r2)
    base_metrics_scores[est_idx, 1] = np.mean(all_fold_mse)
    base_metrics_scores[est_idx, 2] = np.mean(all_fold_mae)

# Evaluate models with PCA
for est_idx, est in enumerate(base_models):
    # Create a pipeline with PCA
    
    # pipline -> https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html
    # preprocessing of data -> https://scikit-learn.org/stable/modules/preprocessing.html 
    # feature scaling -> https://scikit-learn.org/stable/auto_examples/preprocessing/plot_scaling_importance.html#sphx-glr-auto-examples-preprocessing-plot-scaling-importance-py
    # PCA -> https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
    
    pipeline = make_pipeline(StandardScaler(), PCA(n_components=n_components), est)
    
    all_fold_r2 = []
    all_fold_mse = []
    all_fold_mae = []
    for train_idx, test_idx in rskf.split(X, y):
        clf = clone(pipeline)
        clf.fit(X.iloc[train_idx], y[train_idx])
        y_pred = clf.predict(X.iloc[test_idx])

        all_fold_r2.append(r2_score(y[test_idx], y_pred))
        all_fold_mse.append(mean_squared_error(y[test_idx], y_pred))
        all_fold_mae.append(mean_absolute_error(y[test_idx], y_pred))

    pca_metrics_scores[est_idx, 0] = np.mean(all_fold_r2)
    pca_metrics_scores[est_idx, 1] = np.mean(all_fold_mse)
    pca_metrics_scores[est_idx, 2] = np.mean(all_fold_mae)

# Create DataFrames with the model names and their corresponding metrics
base_metrics_df = pd.DataFrame(base_metrics_scores, columns=['R^2 Score', 'MSE', 'MAE'])
base_metrics_df.insert(0, 'Model', model_names)

pca_metrics_df = pd.DataFrame(pca_metrics_scores, columns=['R^2 Score', 'MSE', 'MAE'])
pca_metrics_df.insert(0, 'Model', model_names)

# Display the DataFrames
print("Base Model Metrics:")
print(base_metrics_df)
print("\nPCA Model Metrics:")
print(pca_metrics_df)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Define the parameter grids
param_grid_gbr = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3],
}

param_grid_rfr = {
    'n_estimators': [100, 200, 300],
    'max_features': [None, 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

param_grid_knr = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
}

# Set up the model base with PCA and without PCA
NUM_TRIALS = 1
n_components = 2

# Model base definitions without PCA
model_bases = {
    'RandomForestRegressor': RandomForestRegressor(random_state=42),
    'GradientBoostingRegressor': GradientBoostingRegressor(random_state=42),
    'KNeighborsRegressor': KNeighborsRegressor()
}

# Model base definitions with PCA
model_bases_with_pca = {
    'RandomForestRegressor': make_pipeline(StandardScaler(), PCA(n_components=n_components), RandomForestRegressor(random_state=42)),
    'GradientBoostingRegressor': make_pipeline(StandardScaler(), PCA(n_components=n_components), GradientBoostingRegressor(random_state=42)),
    'KNeighborsRegressor': make_pipeline(StandardScaler(), PCA(n_components=n_components), KNeighborsRegressor())
}

# Parameter grids for each model
param_grids = {
    'RandomForestRegressor': param_grid_rfr,
    'GradientBoostingRegressor': param_grid_gbr,
    'KNeighborsRegressor': param_grid_knr
}

# Arrays to store scores for models without PCA and with PCA
non_nested_scores = {name: np.zeros(NUM_TRIALS) for name in model_bases}
nested_scores = {name: np.zeros(NUM_TRIALS) for name in model_bases}
best_params = {name: [] for name in model_bases}

non_nested_scores_pca = {name: np.zeros(NUM_TRIALS) for name in model_bases}
nested_scores_pca = {name: np.zeros(NUM_TRIALS) for name in model_bases}
best_params_pca = {name: [] for name in model_bases}

# Loop for each trial and each model
for i in range(NUM_TRIALS):
    for name in model_bases:
        # Choose cross-validation techniques for the inner and outer loops
        inner_cv = KFold(n_splits=4, shuffle=True, random_state=i)
        outer_cv = KFold(n_splits=4, shuffle=True, random_state=i)

        # Non-nested parameter search and scoring without PCA
        # nested vs not nested -> https://scikit-learn.org/stable/auto_examples/model_selection/plot_nested_cross_validation_iris.html
        # gridsearchCV -> https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
        # LogisticRegressionCV -> https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegressionCV.html#sklearn.linear_model.LogisticRegressionCV
        clf = GridSearchCV(estimator=model_bases[name], param_grid=param_grids[name], cv=inner_cv, scoring='r2')
        clf.fit(X, y)
        non_nested_scores[name][i] = clf.best_score_
        best_params[name].append(clf.best_params_)

        # Nested CV with parameter optimization without PCA
        nested_score = cross_val_score(clf, X=X, y=y, cv=outer_cv, scoring='r2')
        nested_scores[name][i] = nested_score.mean()

        # Repeat the process with PCA
      
        clf_with_pca = GridSearchCV(estimator=model_bases_with_pca[name], param_grid=param_grids[name], cv=inner_cv, scoring='r2')
        clf_with_pca.fit(X, y)
        non_nested_scores_pca[name][i] = clf_with_pca.best_score_
        best_params_pca[name].append(clf_with_pca.best_params_)

        # Nested CV with parameter optimization with PCA
        nested_score_pca = cross_val_score(clf_with_pca, X=X, y=y, cv=outer_cv, scoring='r2')
        nested_scores_pca[name][i] = nested_score_pca.mean()


# Calculate the difference in scores between the non-nested and nested cross-validation for each model
for name in model_bases:
    score_difference = non_nested_scores[name] - nested_scores[name]
    score_difference_pca = non_nested_scores_pca[name] - nested_scores_pca[name]
    print(f"Model: {name}")
    print(f"Without PCA - Average difference of {score_difference.mean():.6f} with std. dev. of {score_difference.std():.6f}.")
    print(f"With PCA - Average difference of {score_difference_pca.mean():.6f} with std. dev. of {score_difference_pca.std():.6f}.")

    # Best parameters for each trial without PCA
    print("Best parameters without PCA for each trial:", best_params[name])
    # Best parameters for each trial with PCA
    print("Best parameters with PCA for each trial:", best_params_pca[name])

    plt.figure(figsize=(12, 10))

    plt.subplot(221)
    plt.plot(non_nested_scores[name], color='r', label='Non-Nested CV')
    plt.plot(nested_scores[name], color='b', label='Nested CV')
    plt.title(f'Non-Nested vs Nested CV - {name} (without PCA)')
    plt.xlabel('Trial #')
    plt.ylabel('Score')
    plt.legend()

    plt.subplot(222)
    plt.bar(range(NUM_TRIALS), score_difference)
    plt.title(f'Score Difference - {name} (without PCA)')
    plt.xlabel('Trial #')
    plt.ylabel('Score Difference')

    # Plot scores on each trial for nested and non-nested CV for each model with PCA
    plt.subplot(223)
    plt.plot(non_nested_scores_pca[name], color='r', label='Non-Nested CV')
    plt.plot(nested_scores_pca[name], color='b', label='Nested CV')
    plt.title(f'Non-Nested vs Nested CV - {name} (with PCA)')
    plt.xlabel('Trial #')
    plt.ylabel('Score')
    plt.legend()

    plt.subplot(224)
    plt.bar(range(NUM_TRIALS), score_difference_pca)
    plt.title(f'Score Difference - {name} (with PCA)')
    plt.xlabel('Trial #')
    plt.ylabel('Score Difference')

    plt.tight_layout()
    plt.show()
```
